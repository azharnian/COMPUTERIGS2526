## Data Preperation : Normalizatrion - MinMaxScaler sklearn

```python
from sklearn.preprocessing import MinMaxScaler
data = [
    [12_000_000, 33],
    [35_000_000, 45],
    [4_000_000, 23],
    [6_500_000, 26],
    [9_000_000, 29]
]

scaler = MinMaxScaler()
scaler.fit(data)
print(scaler.transform(data))
```

## Data Preperation : Stadarization - Z-Score

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data)
print(scaler.transform(data))
```
